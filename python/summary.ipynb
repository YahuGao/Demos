##### 人工智能的各种方法:
- 机器学习: 人工智能的一个特殊字领域, 其目标是仅靠观察训练数据来自动开发程序(模型). 将数据转换为程序的过程叫做学习.
- 深度学习: 机器学习的众多分支之一, 它的模型是一长串几何函数, 一个接一个的作用在数据上, 这些运算被组织成模块, 叫做层. 深度学习模型通常是层的堆叠. 这些层由权重来参数化, 权重是在训练过程中需要学习的参数. 模型的知识保存在它的权重中, 学习的过程就是为这些权重找到正确的值.
##### 深度学习在机器学习领域的特殊之处
深度学习已经为许多大型科技公司提供了巨大的商业价值, 并且实现了人类水平的语音识别, 智能助理, 人类水平的图像分类, 极大改进的机器翻译等等. 长远来看, 它仍然是一场改变我们经济和生活的重大革命.
##### 如何看待深度学习
"它并不复杂, 只是很多而已" -- 费曼  
深度学习中, 一切皆是向量, 一切都是几何空间中的点. 首先, 将模型输入(文本, 图像等)和目标向量化, 深度学习中的每一层都对通过它的数据做一个简单的几何变换. 模型中的层链共同形成了一个非常复杂的几何变换. 这种几何变换有一个关键的性质, 必须是可微的, 只有这样才能通过梯度下降来学习其参数. 这意味着, 从输入到输出的几何变形必须是平滑且连续的.

##### 机器学习的通用工作流程
- 定义问题: 有哪些数据可用?想要预测什么?是否需要收集更多数据或雇人为数据集手动添加标签?
- 找到能够可靠评估目标成功的方法. 对于简单任务而言, 可以用预测精度, 但很多情况下都需要与领域相关的复杂指标.
- 准备用于评估模型的验证过程. 训练集, 验证集和测试集是必须定义的. 验证集和测试集的标签不应该泄漏到训练数据中去
- 数据向量化. 将数据转换为向量, 并预处理.
- 开发第一个模型, 它要打败基于常识的简单基准方法, 从而表明机器学习能够解决你的问题, 因为事实并非总是如此. 
- 调节超参数和添加正则化来逐步改进模型的架构.
- 调节超参数时要小心验证集的过拟合. 即超参数可能会过于针对验证集而优化. 为了避免该问题, 保留一个单独的测试集.

##### 关键的网络架构
输入模式与适当的网络架构之间的对应关系.
- 向量数据: 密集连接网络
- 图像数据: 二维卷积网络
- 声音数据: 一维卷积神经网络(首选)或循环神经网络.
- 文本数据: 一维卷积神经网络(首选)或循环神经网络.
- 时间序列数据: 循环神经网络(首选)或一维卷积网络
- 其它类型的序列数据: 循环神经网络或一维卷积网络. 如果数据顺序非常重要(比如时间序列, 但文本不是), 那么首选循环神经网络
- 视频数据: 三维卷积网络(如果你需要捕捉运动效果), 或者帧级的二维神经网络(用于特征提取) + 循环神经网络或者一维卷积网络(用于处理得到的序列)
- 立体数据: 三维卷积神经网络.

密集连接网络(Dense层的堆叠, 用于处理向量数据(向量批量). 这种网络假设输入特征中没有特定的结构. 由于Dense层的每个单元都和其它所有单元相连接. 这种层试图映射任意两个输入特征之间的关系, 它与二维卷积层不同, 后者仅查看局部关系
